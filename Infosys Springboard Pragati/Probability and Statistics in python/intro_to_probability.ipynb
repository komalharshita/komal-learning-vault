```python
#set operations
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8} 

#union
C = A.union(B)
print("Union: ", C)
union = A | B
print("Union using | operator: ", union)

#intersection
D = A.intersection(B)
print("Intersection: ", D)

intersection = A & B
print("Intersection using & operator: ", intersection)

#difference
diff = A.difference(B)
print("Difference A - B: ", diff)

dif = A-B
print("Difference using - operator A - B: ", dif)
```

```python
'''Sum rule: If an event can occur in m ways and a second event can occur
    in n ways, and the two events cannot occur at the same time, then the total
    number of ways either event can occur is m+n'''
```

```python
'''Bayes Theorem: Describes the probability of an event, 
based on prior knowledge'''
```

```python
'''
eg: 1% people have a disease
p(D) = 0.01
Test is 90% accurate when disease is present
P(+|D) = 0.9
Test is 95% accurate when disease is absent
P(-|D') = 0.95
Find false positive rate
P(+|D') = 0.05
Find total probability of positive test
P(+) = P(+|D) * P(D) + P(+|D') * P(D')
P(+) = 0.9 * 0.01 + 0.05 * 0.99 = 0.0595
Bayes theorem:
P(D|+) = P(+|D) * P(D) / P(+)
P(D|+) = (0.9 * 0.01) / 0.05
P(D|+) = 0.18
'''
```
